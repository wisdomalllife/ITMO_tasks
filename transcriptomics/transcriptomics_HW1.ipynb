#DESeq

```r
samples <- read_tsv("/content/phenotable.tsv")
files <- file.path(samples$path, "abundance.h5")  # path to abundance files in kallisto dirs
txi <- tximport(files, type = 'kallisto', txOut = T)
```

```r
# Create DESeq object
ddsTxi <- DESeqDataSetFromTximport(txi,
                                   colData = samples,
                                   design = ~ condition)  # change design if needed
# note that ~ condition*time is equal to ~ condition + time + condition:time

# filter low-expressed genes
dds <- ddsTxi[rowSums(counts(ddsTxi)) >= 10,]

# relevel condition
dds$condition <- relevel(dds$condition, ref = "Control")
```

```r
# Calculate DE
dds <- DESeq(dds)

# another way of doing the same
# dds <- ddsTxi %>% estimateSizeFactors %>% estimateDispersions %>% nbinomWaldTest


# variance-stabilizing transformation of data (important for clustering or ML)
vsd <- varianceStabilizingTransformation(dds)
```

```r
# Plot a heatmap of 100 most expressed genes
select <- order(rowMeans(counts(dds,normalized=TRUE)),
                decreasing=TRUE)[1:30]
df <- as.data.frame(colData(dds)[,"condition"])
pheatmap(assay(vsd)[select,], cluster_rows=T, show_rownames=T,
         cluster_cols=T, annotation_col=df)
```

```r
# Distance between samples heatmap
sampleDists <- dist(t(assay(vsd)))

sampleDistMatrix <- as.matrix(sampleDists)
rownames(sampleDistMatrix) <- paste(vsd$condition, sep="-")
colnames(sampleDistMatrix) <- paste(vsd$condition, sep="-")
colors <- colorRampPalette( rev(brewer.pal(9, "Blues")) )(255)
pheatmap(sampleDistMatrix,
         clustering_distance_rows=sampleDists,
         clustering_distance_cols=sampleDists,
         color = colors)
```

# GSEA

```r
# GSEA
pathways <- fgsea::gmtPathways("/content/wikipathways-20240310-gmt-Mus_musculus.gmt")
```

```r
translated_names <- res %>%
  as.data.frame() %>%
  na.omit() %>%
  rownames_to_column("ensembl_transcript_id_version")

res2 <- getBM(attributes = c('ensembl_gene_id','ensembl_transcript_id_version','entrezgene_id'), filters = 'ensembl_transcript_id_version', values = translated_names$ensembl_transcript_id_version, mart = mart)
translated_names <- merge(translated_names, res2, by = "ensembl_transcript_id_version")
translated_names <- translated_names %>%
  #pull(ens_id) %>%
  #bitr("ENSEMBL", c("GENENAME", "ENTREZID"), "org.Mm.eg.db") %>%
  distinct(ensembl_transcript_id_version, .keep_all = T)
```

```r
ranks_for_gsea <- translated_names %>%
  arrange(desc(stat)) %>%
  mutate(NAME = ifelse(is.na(entrezgene_id), ensembl_gene_id, entrezgene_id)) %>%
  transmute(NAME, stat) %>%
  deframe()
```

```r
# run fgsea
fgsea_results <- fgsea::fgseaMultilevel(pathways, ranks_for_gsea)

# results
head(fgsea_results[order(pval), ])
```

```r
# GSEA plot for one pathway
fgsea::plotEnrichment(pathways[["mRNA processing%WikiPathways_20240310%WP310%Mus musculus"]], ranks_for_gsea) + labs(title="mRNA processing%WikiPathways_20240310%WP310%Mus musculus")
```

```r
# top and bottom 10 pathways
topPathwaysUp <- fgsea_results[ES > 0][head(order(pval), n=10), pathway]
topPathwaysDown <- fgsea_results[ES < 0][head(order(pval), n=10), pathway]
topPathways <- c(topPathwaysUp, rev(topPathwaysDown))
fgsea::plotGseaTable(pathways[topPathways], ranks_for_gsea, fgsea_results,
                     gseaParam=0.5)
```
